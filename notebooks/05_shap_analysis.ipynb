# Phase 7: SHAP Analysis
Explainability of the XGBoost Model.

```python
import pandas as pd
import xgboost as xgb
import shap
import pickle
import matplotlib.pyplot as plt

# Load model
with open('../model/fraud_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Load data
df = pd.read_csv('../data/processed/train_sample_processed.csv')
X = df.drop(columns=['isFraud'])

# Sample for SHAP (it's slow on full data)
X_sample = X.sample(1000, random_state=42)

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_sample)

# Summary plot
shap.summary_plot(shap_values, X_sample, show=False)
plt.savefig('../reports/shap_summary.png', bbox_inches='tight')
plt.close()

# Feature Importance from model
xgb.plot_importance(model, max_num_features=20)
plt.savefig('../reports/feature_importance.png', bbox_inches='tight')
plt.close()

print("SHAP analysis complete. Plots saved in reports/")
```
